```python
import tkinter # import the Tkinter library for creating graphical interfaces.
from tkinter import  *
# create the root window for the calculator.
root=Tk() 
root.title("simple calculator") 
root.geometry("570x600+100+200") 
root.resizable(False,False)     
root.configure(bg="grey")      

equation = ""



#  function to update the equation shown on the screen when a button is pressed 
def show(value):
    global equation # access the global equation variable
    equation+=value # append the pressed value to the current equation
    label_result.config(text=equation) # update the label to display the new equation

# function to clear the equation and reset the display .
def clear():
    global equation # access the global equation.
    equation = ""    # Reset the equation to an empty string.
    label_result.config(text=equation) # clear the display label.

# function to calculate the result of the equation.

def calculate():
    global equation # access the global equation variable.
    result =""      # initialize an empty string for the result
    if equation !="":  # if equation is not empty 
        try:
            result= eval(equation)  # evaluate the equation and store the result.

        except:
            result ="error"   # handle error.( eg - invalid syntax.), show "error"
            equation = ""   # reset the equation to an empty string

    label_result.config(text=result)   # display the result on the screen .
    
label_result= Label(root,width=25,height=2,text="",font=("arial",40,"bold"), fg="orange" ,bg="white")
label_result.pack()

Button(root,text="c",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="white",bg="blue",command=lambda:clear()).place(x=10,y=100)
Button(root,text="/",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="white",  bg="blue",command=lambda:show("/")).place(x=150,y=100)
Button(root,text="%",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="white",bg="blue",command=lambda:show("%")).place(x=290,y=100)
Button(root,text="*",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="white",bg="blue",command=lambda:show("*")).place(x=430,y=100)


Button(root,text="7",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="#fff", bg="#2a2d36",command=lambda:show("7")).place(x=10,y=200)
Button(root,text="8",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="#fff",bg="#2a2d36",command=lambda:show("8")).place(x=150,y=200)
Button(root,text="9",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="#fff",bg="#2a2d36",command=lambda:show("9")).place(x=290,y=200)
Button(root,text="-",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="white",bg="blue",command=lambda:show("-")).place(x=430,y=200)

Button(root,text="4",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="#fff",bg="#2a2d36",command=lambda:show("4")).place(x=10,y=300)
Button(root,text="5",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="#fff",bg="#2a2d36",command=lambda:show("5")).place(x=150,y=300)
Button(root,text="6",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="#fff",bg="#2a2d36",command=lambda:show("6")).place(x=290,y=300)
Button(root,text="+",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="white",bg="blue",command=lambda:show("+")).place(x=430,y=300)

Button(root,text="1",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="#fff",bg="#2a2d36",command=lambda:show("1")).place(x=10,y=400)
Button(root,text="2",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="#fff",bg="#2a2d36",command=lambda:show("2")).place(x=150,y=400)
Button(root,text="3",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="#fff",bg="#2a2d36",command=lambda:show("3")).place(x=290,y=400)
Button(root,text="0",width=11,height=1,font=("arial",30,"bold"),bd=3,fg="#fff",bg="#2a2d36",command=lambda:show("0")).place(x=10,y=500)

Button(root,text=".",width=5,height=1,font=("arial",30,"bold"), bd=3,fg="white",bg="#2a2d36",command=lambda:show(".")).place(x=290,y=500)
Button(root,text="=",width=5,height=3,font=("arial",30,"bold"), bd=3,fg="#fff",bg="blue",command=lambda:calculate()).place(x=430,y=400)

root.mainloop()
 
```
